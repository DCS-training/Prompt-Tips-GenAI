Retrieval Augmented Generation (RAG), a common LLM application that retrieves contextual documents from an external dataset,

My focus is about giving a short demo for a specific task. For these type of platforms, we have various options (including but most likely not limited to followings)

- **[Langchain](https://python.langchain.com/docs/get_started/introduction)** is a framework for developing applications powered by language models. It enables applications that are context aware!

- **[Embedchain](https://docs.embedchain.ai/get-started/quickstart)** is a Data Platform for LLMs - load, index, retrieve, and sync any unstructured data. Using embedchain, you can easily create LLM powered apps over any data.

- **[LlamaIndex](https://www.llamaindex.ai/)** is a simple, flexible data framework for connecting
custom data sources to large language models.

Now we import the dependencies:

```python
# This module provides a portable way of using operating system dependent functionality.
import os

#Simply import App functionality
from embedchain import App
```

We instanciate the embechain bot. Remember to change the API key with your OpenAI api key.

```python
# Using OpenAI API Key for the connection
# Without your key, the rest is not executable indeed!
os.environ["OPENAI_API_KEY"] = "YOUR-OPENAI-API-KEY"

# Initiation of the bot
CDCS_bot = App()
```

Now, add different data sources using embedchain's `.add()` method:

```python
CDCS_bot.add("web_page", "https://www.cdcs.ed.ac.uk/news/cdcs-training-programme-semester-2-2024")
# Let us say goal is to look at training events
```

*Your* bot is ready now. Ask your bot any questions using `.query()` method:

```python
CDCS_bot.query("Give me the list of trainings related to R programming?.")
```

```python
CDCS_bot.query("Give me the list of trainings related to R programming?. Create the output as a bullet list in the form of markdown")
```

```python
CDCS_bot.query("Give me the list of trainings related to R programming?. Create the output in a Jsonl file format")
```
